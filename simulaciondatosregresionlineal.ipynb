```python
## Simulacion de datos para regresion lineal
# R2
# EAM

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error
import matplotlib.pyplot as plt

# --- 1. Generación de Datos Simulados para Regresión ---
print("--- 1. Generación de Datos Simulados para Regresión ---")

# Número de muestras
n_samples = 100

# Generar una característica (X)
np.random.seed(42) # Para reproducibilidad
X_simulada = 10 * np.random.rand(n_samples, 1)

# Generar la variable objetivo (y) con una relación lineal y ruido
# y = 2 * X + 5 + ruido_aleatorio
y_real = 2 * X_simulada + 5 + np.random.randn(n_samples, 1) * 2

# Convertir a DataFrame para consistencia
df_simulado = pd.DataFrame({'X_simulada': X_simulada.flatten(), 'y_real': y_real.flatten()})

print("DataFrame de datos simulados creado:")
print(df_simulado.head())
print(f"Dimensiones de los datos simulados: {df_simulado.shape}")

# --- 2. División de Datos (Entrenamiento y Prueba) ---
print("\n--- 2. División de Datos ---")

# Para regresión, no necesitamos stratify a menos que y sea categórica, lo cual no es el caso aquí.
X_train_sim, X_test_sim, y_train_sim, y_test_sim = train_test_split(
    X_simulada, y_real, test_size=0.2, random_state=42
)

print(f"Dimensiones del conjunto de entrenamiento: X={X_train_sim.shape}, y={y_train_sim.shape}")
print(f"Dimensiones del conjunto de prueba: X={X_test_sim.shape}, y={y_test_sim.shape}")

# --- 3. Entrenamiento del Modelo de Regresión ---
print("\n--- 3. Entrenamiento del Modelo de Regresión ---")

model_regresion = LinearRegression()
model_regresion.fit(X_train_sim, y_train_sim)

print("Modelo de Regresión Lineal entrenado.")
print(f"Coeficiente (pendiente): {model_regresion.coef_[0][0]:.2f}")
print(f"Intercepto: {model_regresion.intercept_[0]:.2f}")

# --- 4. Realizar Predicciones en el Conjunto de Prueba ---
print("\n--- 4. Realizar Predicciones ---")
y_pred_sim = model_regresion.predict(X_test_sim)

# --- 5. Evaluación de Métricas de Regresión (R2 y EAM) ---
print("\n--- 5. Evaluación de Métricas de Regresión (R² y EAM) ---")

r2 = r2_score(y_test_sim, y_pred_sim)
eam = mean_absolute_error(y_test_sim, y_pred_sim)

print(f"Métrica R² (Coeficiente de Determinación): {r2:.4f}")
print(f"Métrica EAM (Error Absoluto Medio): {eam:.4f}")

# --- 6. Visualización de los Resultados ---
print("\n--- 6. Visualización de los Resultados ---")
plt.figure(figsize=(10, 6))
plt.scatter(X_test_sim, y_test_sim, color='blue', label='Valores Reales de Prueba')
plt.plot(X_test_sim, y_pred_sim, color='red', label='Predicciones del Modelo', linewidth=2)
plt.title(f'Resultados de la Regresión (R²: {r2:.2f}, EAM: {eam:.2f})')
plt.xlabel('X Simulada')
plt.ylabel('y Real / Predicha')
plt.legend()
plt.grid(True)
plt.show()

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error
import matplotlib.pyplot as plt

# --- 1. Generación de Datos Simulados para Regresión ---
print("--- 1. Generación de Datos Simulados para Regresión ---")

# Número de muestras
n_samples = 100

# Generar una característica (X)
np.random.seed(42) # Para reproducibilidad
X_simulada = 10 * np.random.rand(n_samples, 1)

# Generar la variable objetivo (y) con una relación lineal y ruido
# y = 2 * X + 5 + ruido_aleatorio
y_real = 2 * X_simulada + 5 + np.random.randn(n_samples, 1) * 2

# Convertir a DataFrame para consistencia
df_simulado = pd.DataFrame({'X_simulada': X_simulada.flatten(), 'y_real': y_real.flatten()})

print("DataFrame de datos simulados creado:")
print(df_simulado.head())
print(f"Dimensiones de los datos simulados: {df_simulado.shape}")

# --- 2. División de Datos (Entrenamiento y Prueba) ---
print("\n--- 2. División de Datos ---")

# Para regresión, no necesitamos stratify a menos que y sea categórica, lo cual no es el caso aquí.
X_train_sim, X_test_sim, y_train_sim, y_test_sim = train_test_split(
    X_simulada, y_real, test_size=0.2, random_state=42
)

print(f"Dimensiones del conjunto de entrenamiento: X={X_train_sim.shape}, y={y_train_sim.shape}")
print(f"Dimensiones del conjunto de prueba: X={X_test_sim.shape}, y={y_test_sim.shape}")

# --- 3. Entrenamiento del Modelo de Regresión ---
print("\n--- 3. Entrenamiento del Modelo de Regresión ---")

model_regresion = LinearRegression()
model_regresion.fit(X_train_sim, y_train_sim)

print("Modelo de Regresión Lineal entrenado.")
print(f"Coeficiente (pendiente): {model_regresion.coef_[0][0]:.2f}")
print(f"Intercepto: {model_regresion.intercept_[0]:.2f}")

# --- 4. Realizar Predicciones en el Conjunto de Prueba ---
print("\n--- 4. Realizar Predicciones ---")
y_pred_sim = model_regresion.predict(X_test_sim)

# --- 5. Evaluación de Métricas de Regresión (R2 y EAM) ---
print("\n--- 5. Evaluación de Métricas de Regresión (R² y EAM) ---")

r2 = r2_score(y_test_sim, y_pred_sim)
eam = mean_absolute_error(y_test_sim, y_pred_sim)

print(f"Métrica R² (Coeficiente de Determinación): {r2:.4f}")
print(f"Métrica EAM (Error Absoluto Medio): {eam:.4f}")

# --- 6. Visualización de los Resultados ---
print("\n--- 6. Visualización de los Resultados ---")
plt.figure(figsize=(10, 6))
plt.scatter(X_test_sim, y_test_sim, color='blue', label='Valores Reales de Prueba')
plt.plot(X_test_sim, y_pred_sim, color='red', label='Predicciones del Modelo', linewidth=2)
plt.title(f'Resultados de la Regresión (R²: {r2:.2f}, EAM: {eam:.2f})')
plt.xlabel('X Simulada')
plt.ylabel('y Real / Predicha')
plt.legend()
plt.grid(True)
plt.show()

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error
import matplotlib.pyplot as plt

# --- 1. Generación de Datos Simulados para Regresión ---
print("--- 1. Generación de Datos Simulados para Regresión ---")

# Número de muestras
n_samples = 100

# Generar una característica (X)
np.random.seed(42) # Para reproducibilidad
X_simulada = 10 * np.random.rand(n_samples, 1)

# Generar la variable objetivo (y) con una relación lineal y ruido
# y = 2 * X + 5 + ruido_aleatorio
y_real = 2 * X_simulada + 5 + np.random.randn(n_samples, 1) * 2

# Convertir a DataFrame para consistencia
df_simulado = pd.DataFrame({'X_simulada': X_simulada.flatten(), 'y_real': y_real.flatten()})

print("DataFrame de datos simulados creado:")
print(df_simulado.head())
print(f"Dimensiones de los datos simulados: {df_simulado.shape}")

# --- 2. División de Datos (Entrenamiento y Prueba) ---
print("\n--- 2. División de Datos ---")

# Para regresión, no necesitamos stratify a menos que y sea categórica, lo cual no es el caso aquí.
X_train_sim, X_test_sim, y_train_sim, y_test_sim = train_test_split(
    X_simulada, y_real, test_size=0.2, random_state=42
)

print(f"Dimensiones del conjunto de entrenamiento: X={X_train_sim.shape}, y={y_train_sim.shape}")
print(f"Dimensiones del conjunto de prueba: X={X_test_sim.shape}, y={y_test_sim.shape}")

# --- 3. Entrenamiento del Modelo de Regresión ---
print("\n--- 3. Entrenamiento del Modelo de Regresión ---")

model_regresion = LinearRegression()
model_regresion.fit(X_train_sim, y_train_sim)

print("Modelo de Regresión Lineal entrenado.")
print(f"Coeficiente (pendiente): {model_regresion.coef_[0][0]:.2f}")
print(f"Intercepto: {model_regresion.intercept_[0]:.2f}")

# --- 4. Realizar Predicciones en el Conjunto de Prueba ---
print("\n--- 4. Realizar Predicciones ---")
y_pred_sim = model_regresion.predict(X_test_sim)

# --- 5. Evaluación de Métricas de Regresión (R2 y EAM) ---
print("\n--- 5. Evaluación de Métricas de Regresión (R² y EAM) ---")

r2 = r2_score(y_test_sim, y_pred_sim)
eam = mean_absolute_error(y_test_sim, y_pred_sim)

print(f"Métrica R² (Coeficiente de Determinación): {r2:.4f}")
print(f"Métrica EAM (Error Absoluto Medio): {eam:.4f}")

# --- 6. Visualización de los Resultados ---
print("\n--- 6. Visualización de los Resultados ---")
plt.figure(figsize=(10, 6))
plt.scatter(X_test_sim, y_test_sim, color='blue', label='Valores Reales de Prueba')
plt.plot(X_test_sim, y_pred_sim, color='red', label='Predicciones del Modelo', linewidth=2)
plt.title(f'Resultados de la Regresión (R²: {r2:.2f}, EAM: {eam:.2f})')
plt.xlabel('X Simulada')
plt.ylabel('y Real / Predicha')
plt.legend()
plt.grid(True)
plt.show()
```
